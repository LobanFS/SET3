```python
import pandas as pd
import matplotlib.pyplot as plt

df_merge = pd.read_csv("merge_results.csv")
plt.figure()
for t in ["random", "reverse", "almost"]:
    sub = df_merge[df_merge["type"] == t]
    plt.plot(sub["size"], sub["ms"], label=t)

plt.xlabel("Размер массива")
plt.ylabel("Время, мс (среднее из 5 запусков)")
plt.title("MERGE SORT: зависимости времени от размера массива")
plt.grid(True)
plt.legend()
plt.show()

for t in ["random", "reverse", "almost"]:
    sub = df_merge[df_merge["type"] == t]
    plt.figure()
    plt.plot(sub["size"], sub["ms"])
    plt.xlabel("Размер массива")
    plt.ylabel("Время, мс")
    plt.title(f"MERGE SORT: {t}")
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_hybrid = pd.read_csv("hybrid_results.csv")

plt.figure()
for t in ["random", "reverse", "almost"]:
    sub = df_hybrid[df_hybrid["type"] == t]
    plt.plot(sub["size"], sub["ms"], label=t)

plt.xlabel("Размер массива")
plt.ylabel("Время, мс")
plt.title("HYBRID: зависимости времени от размера массива")
plt.grid(True)
plt.legend()
plt.show()

for t in ["random", "reverse", "almost"]:
    sub = df_hybrid[df_hybrid["type"] == t]
    plt.figure()
    plt.plot(sub["size"], sub["ms"])
    plt.xlabel("Размер массива")
    plt.ylabel("Время, мс")
    plt.title(f"HYBRID SORT: {t}")
    plt.grid(True)
    plt.show()
```

```python
types = ["random", "reverse", "almost"]
for t in types:
    m = df_merge[df_merge["type"] == t]
    h = df_hybrid[df_hybrid["type"] == t]
    plt.figure()
    plt.plot(m["size"], m["ms"], label="MergeSort", linewidth=1.5)
    plt.plot(h["size"], h["ms"], label="Hybrid", linewidth=1.5)

    plt.title(f"{t}: Merge vs Hybrid")
    plt.xlabel("Размер массива")
    plt.ylabel("Время (ms)")
    plt.grid(True)
    plt.legend()
    plt.show()
```
